# `run_metrics.ndjson` exploration

Append-only JSON Lines in each run folder: `logs/<run>/run_metrics.ndjson` (or inside archived `*.zip` as `<run>/run_metrics.ndjson`).

Record types: `state` (with `vm_summary`) and `ai_decision` (tokens, latency, `status`).

API equivalent: `GET /api/runs/{run}/metrics?summary=1` from the control dashboard.

```python
from pathlib import Path

import pandas as pd

REPO = Path("..").resolve()
RUN = "2026-03-29-10-11"  # change to your run folder name
path = REPO / "logs" / RUN / "run_metrics.ndjson"
df = pd.read_json(path, lines=True)
df["type"].value_counts()
```

```python
states = df[df["type"] == "state"].copy()
ai = df[df["type"] == "ai_decision"].copy()
executed = ai[ai["status"] == "executed"]
states[["event_index", "timestamp"]].head(), executed[["event_index", "total_tokens", "latency_ms", "status"]].head()
```
